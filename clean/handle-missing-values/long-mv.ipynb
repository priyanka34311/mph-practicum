```python
# Set up environment

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Read in data

long = pd.read_csv('long-term.csv', na_values = ' ')
```

```python
# Check to ensure that null values are detected
long.info()
```

```python
# Check for missing values in Q15
long['Q15'].value_counts().sum()
long['Q15'] = long['Q15'].fillna(-999)
long['Q15'].value_counts()
```

```python
# For Q16 & Q16_333_TEXT - 
#everyone with Q15 = 1 should be not applicable

long.loc[long['Q15']==1,['Q16','Q16_333_TEXT']] = -888
```

```python
# Check to make sure above worked
long.loc[long['Q15']==1,['Q16','Q16_333_TEXT']]
```

```python
# Note all the missing values in Q15 as MV in Q16+string 
long.loc[long['Q15']==-999,['Q16','Q16_333_TEXT']] = -999
long.loc[long['Q15']==-999,['Q16','Q16_333_TEXT']]
```

```python
long.loc[long['Q15']==0,['Q16','Q16_333_TEXT']] # From this we see 261 
# should be marked as missing. The rest will be n/a as -888

long.loc[261,['Q16','Q16_333_TEXT']] = -999

long.loc[long['Q15']==0,['Q16','Q16_333_TEXT']]
```

```python
# Handle the rest of Q16 string variable
long['Q16_333_TEXT'].value_counts()
long['Q16_333_TEXT'] = long['Q16_333_TEXT'].fillna(-888)
long['Q16_333_TEXT'].value_counts()
```

```python
# Check before filling in the rest of the mv

long.to_csv('check.csv',index=False)
```

```python
# Fill in the rest as MV -999

long = long.fillna(-999)
```
